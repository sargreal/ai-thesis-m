```python
import pandas
import numpy as np

# df =  pandas.read_csv('eval_htn.csv')
df =  pandas.read_csv('evaluation_robustness.csv')

df.columns = df.columns.str.strip()
df = df.applymap(lambda x: " ".join(x.split()) if isinstance(x, str) else x)

df.replace('dfs:ord-dec:soca', "default", inplace=True)
df.replace('dfs:makespan:ord-dec:soca', "2=makespan", inplace=True)
df.replace('dfs:ord-dec:makespan:soca', "3=makespan", inplace=True)
df.replace('dfs:robust:ord-dec:soca', "2=robustness", inplace=True)
df.replace('dfs:ord-dec:robust:soca', "3=robustness", inplace=True)

df['runtime'] = pandas.to_numeric(df['runtime'], errors='coerce')
df['sol-depth'] = pandas.to_numeric(df['sol-depth'], errors='coerce')
df['makespan'] = pandas.to_numeric(df['makespan'], errors='coerce')
df = df.dropna()
df.head()

```

```python
# grouped = df.groupby('anml-file', dropna=False)
grouped = df.groupby(['anml-file', 'plan-sel'], dropna=False)

grouped.mean(numeric_only=True)
```
